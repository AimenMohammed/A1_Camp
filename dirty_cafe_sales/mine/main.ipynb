```python
import pandas as pd

# 1. إنشاء جدول الطلاب المسجلين (ydf)
ydf = pd.DataFrame({
    'student_id': [101, 102, 103, 104],
    'name': ['Ali', 'Sara', 'Omar', 'Huda']
})

# 2. إنشاء جدول الطلاب الذين حضروا الامتحان (zdf)
zdf = pd.DataFrame({
    'student_id': [102, 104],
    'score': [85, 92]
})

# 3. تطبيق العملية لاستخراج "الغائبين" (الموجودين في ydf وليس zdf)
absentees = pd.merge(ydf, zdf, on='student_id', how='outer', indicator=True) \
              .query('_merge == "left_only"') \
              .drop(columns=['_merge'])

print("الطلاب الذين سجلوا ولم يحضروا الامتحان:")
print(absentees)
```
