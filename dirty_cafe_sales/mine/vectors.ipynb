```python
import math
import re

def get_automatic_bag(texts):
    # دمج النصوص وتنظيفها واستخراج الكلمات الفريدة
    combined_text = " ".join(texts).lower()
    # استخدام re لإزالة علامات الترقيم وبقاء الكلمات فقط
    words = re.findall(r'\w+', combined_text)
    # set تعطينا الكلمات بدون تكرار
    return sorted(list(set(words)))

def text_to_vector(text, bag_of_words):
    words_in_text = re.findall(r'\w+', text.lower())
    return [words_in_text.count(word) for word in bag_of_words]

def calculate_similarity(v1, v2):
    dot_product = sum(a * b for a, b in zip(v1, v2))
    mag1 = math.sqrt(sum(a**2 for a in v1))
    mag2 = math.sqrt(sum(b**2 for b in v2))
    return (dot_product / (mag1 * mag2)) if mag1 and mag2 else 0

# --- التنفيذ ---
text1 = input("أدخل النص الأول: ")
text2 = input("أدخل النص الثاني: ")

# إنشاء القاموس تلقائياً من النصين
auto_bag = get_automatic_bag([text1, text2])

# تحويل النصوص لمتجهات بناءً على القاموس التلقائي
vec1 = text_to_vector(text1, auto_bag)
vec2 = text_to_vector(text2, auto_bag)

print("\n--- النتائج ---")
print(f"القاموس المستخرج تلقائياً ({len(auto_bag)} كلمة):")
print(auto_bag)
print(f"\nVector 1: {vec1}")
print(f"Vector 2: {vec2}")
print(f"نسبة التشابه: {calculate_similarity(vec1, vec2) * 100:.2f}%")
```
